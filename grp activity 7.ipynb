#### 4.Carry out correlation analysis on the dataset and find the correlation factor between (a) ph and Chloramines (b) ph and Conductivity. What do we infer from this correlation value?

```python
co_matrix=data.corr()
plt.subplots(figsize=(10,4))
sns.heatmap(co_matrix,vmin=-.2,vmax=.2,annot=True,linewidth=0.2,cmap='YlGnBu')
```

#### Insights: As inferred from the heat map, the correlation factor between ph and Chloramines is -0.025 which indicates a low negative correlation. Also the correlation factor between ph and Conductivity is 0.014 which indicates a low positive correlation.
